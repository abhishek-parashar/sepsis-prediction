```python
import os
import glob
import pandas
def concatenate(indir = "E:\\Yukti\\Competitions\\SIH 2020\\DATA\\training", outfile = "E:\\Yukti\\Competitions\\SIH 2020\\DATA\\Training Data.csv"):
    os.chdir(indir)
    fileList = glob.glob("*.psv")
    dfList = []
    for filename in fileList:
        print(filename)
        df = pandas.read_csv(filename, delimiter = '|')
        df.drop(df.index[0], inplace = True)
        dfList.append(df)
        concatDf = pandas.concat(dfList, axis=0)
        concatDf.to_csv(outfile)
```
